# Q1. book 

```python
import pandas as pd
import numpy as np
```

```python
data= pd.read_csv('book.csv')
```

```python
data.head()
```

```python
data.shape
```

```python
data.info()
```

```python
#cheking for support on deifrent values 
```

```python
from mlxtend.frequent_patterns import apriori,association_rules
```

```python
test1=apriori(data,min_support=0.5,use_colnames=True)
test1
```

```python
test2=apriori(data,min_support=0.3,use_colnames=True)
test2
```

```python
test3=apriori(data,min_support=0.2,use_colnames=True)
test3
```

```python
# here we found good support (0.1)
```

```python
test4=apriori(data,min_support=0.1,use_colnames=True)
test4
```

```python
freq=association_rules(test4,metric='lift',min_threshold=1)
freq
```

```python
freq.sort_values('lift',ascending=False)[0:20]
```

```python
freq1=association_rules(test4,metric='confidence',min_threshold=0.8)
freq1
```

```python
freq.sort_values('confidence',ascending=False)[0:20]
```

# Q2. movie 

```python
movie=pd.read_csv('my_movies.csv')
movie.head()
```

```python
movie.shape
```

```python
movie.info()
```

```python
# here we use one hot encoding method for categorical data
```

```python
movie1= pd.get_dummies(movie)
movie1
```

```python
# cheking for good support 
```

```python
from mlxtend.frequent_patterns import apriori,association_rules
```

```python
exa=apriori(movie1,min_support=0.2,use_colnames=True)
exa
```

```python
exa1=apriori(movie1,min_support=0.5,use_colnames=True)
exa1
```

```python
# we got 0.5 as a good support 
```

```python
freq=association_rules(exa1,metric='lift',min_threshold=1)
freq
```

```python
freq[freq['lift']>=1.5]
```

```python
freq.sort_values('lift',ascending=False)[0:20]
```

```python
freq1=association_rules(exa,metric='confidence',min_threshold=1)
freq1
```

```python
freq.sort_values('confidence',ascending=False)[0:20]
```
